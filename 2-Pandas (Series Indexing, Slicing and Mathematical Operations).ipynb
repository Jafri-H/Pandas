#### Fetching Values from Panda Series:

```python
import pandas as pd
```

```python
series = pd.Series([1,2,3,4,5], index=('a', 'b', 'c', 'd', 'e'))
print(series)
```

```python
series[0]
```

```python
series [1]
```

#### Slicing Values from Panda Series:

```python
print(series)
```

```python
series [0:3] # ending index is excluded.
```

#### Mathematic Operations on Series:

```python
max(series)
```

```python
min(series)
```

```python
series[series>3]
```

```python
series[series<3]
```

```python
s1 = pd.Series((10,20,30,40))
print(s1)
```

```python
s2 = pd.Series((50,60,70,800))
print(s2)
```

```python
s3 = s1+s2
print(s3)
```

```python
s4 = s1-s2
print(s4)
```
