```python
# Impoorting Pandas
import pandas as pd
```

```python
# Checking Pandas version
pd.__version__
```

```python
list_s = [1, 2, -3, 6.2, 'data_values']
print (list_s)
```

### Creating Pandas Series:

```python
series1 = pd.Series(list_s)
print (series1)
```

```python
print(type(series1))
```

#### Creating Series with Scalar(Single) value:

```python
series= pd.Series(0.5)
print(series)
print(type(series))
```

```python
series0= pd.Series(0.5, index =['a','b','c'])
print(series0)
print(type(series0))
```

#### Creating Series with Tuple

```python
series1 = pd.Series((1,2,3,4))
print(series1)
print(type(series1))
```

#### Creating Series with List
- pd.Series , `S` should be capital in Series.

```python
series1 = pd.Series([1,2,3,4])
print(series1)
print(type(series1))
```

#### Creating Series witth Dictionary:

```python
dict_s = pd.Series({'name':'Jafri' , 'grade':'A1'})
print(dict_s)
print(type(dict_s))
```

#### Creating Empty Series:

```python
series_empty = pd.Series([])
print(series_empty)
print(type(series_empty))
```

#### Changing Default Index:

```python
series3 = pd.Series([1,2,3,4])
print(series3)
```

```python
 # index value count should be same as series value count otherwise 'Value Error'.
series4 = pd.Series([1,2,3,4], index = (['a','b','c', 'd']))
print(series4)
```

#### Changing Datatype:

```python
series4 = pd.Series ([1,2,3,4], index = ['a','b','c','d'], dtype = float)
print(series4)
```

#### Giving Names to Datavalues in Series:

```python
series5 = pd.Series ([1,2,3,4], index = ['a','b','c','d'], dtype = float, name = 'data value' )
print(series5)
```
