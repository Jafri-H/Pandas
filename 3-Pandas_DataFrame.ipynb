#### Creating DataFrame with Dictionaries of Series

```python
dict_ser = {'ID': pd.Series([1,2,3]), 'SN': pd.Series([111,222,333])}
print (dict_ser)
```

```python
df7 = pd.DataFrame(dict_ser)
df7
```

```python
# ❌ Example 1: Dictionary of lists with unequal lengths

# Each key (1D and SN) becomes a column name, and each list becomes that column's data
# '1D' → [11, 22, 33, 44]  (4 values)
# 'SN' → [1, 2, 3]         (3 values)
# pandas requires all columns to have the SAME number of rows
# So this will raise: ValueError: All arrays must be of the same length

# df5 = pd.DataFrame({'1D': [11, 22, 33, 44], 'SN': [1, 2, 3]})
# Result:
#   1D   SN
#  11    1
#  22    2
#  33    3
#  44    ?   --> ❌ Error (missing value, unequal column length)


# ✅ Example 2: List of dictionaries (each dictionary = one row)
# Missing keys are automatically filled with NaN
# Columns are taken as the union of all dictionary keys

# df5 = pd.DataFrame([{'a': 1, 'b': 2}, {'a': 3, 'b': 4, 'c': 5}])
# Result:
#      a    b    c
#  0   1    2   NaN
#  1   3    4   5.0
#
# Explanation:
# Row 1 → {'a':1, 'b':2}  → no 'c' key → NaN
# Row 2 → {'a':3, 'b':4, 'c':5} → all keys present
```
